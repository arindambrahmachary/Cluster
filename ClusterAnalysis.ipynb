# Visualization

```python
# prompt: show me code for histogram for my whole data

df.hist(figsize=(15, 15))
plt.show()
```

```python
# prompt: show  me the code for boxplot

# Boxplot for all numerical features
plt.figure(figsize=(8, 6))
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.show()

# Example boxplot for a specific feature
plt.figure(figsize=(6, 4))
sns.boxplot(x='Alcohol', data=df)
plt.show()
```

```python
# desity plot for whole data

# Density plot for all numerical features
for column in df.select_dtypes(include=np.number).columns:
    plt.figure(figsize=(3, 2))
    sns.kdeplot(df[column])
    plt.title(f'Density Plot of {column}')
    plt.show()
```

```python


# Correlation matrix
correlation_matrix = df.corr()
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
#Checking highest  Relationship between

# Find the highest correlation between features
highest_correlation = correlation_matrix[correlation_matrix != 1].max().max()
row, col = correlation_matrix[correlation_matrix != 1].stack().idxmax()

print(f"The highest correlation is between '{row}' and '{col}' with a value of {highest_correlation:.2f}")
```

```python
#  relation between 'Total_Phenols' and 'Flavanoids'

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Total_Phenols', y='Flavanoids', data=df)
plt.title('Relationship between Total Phenols and Flavanoids')
plt.xlabel('Total Phenols')
plt.ylabel('Flavanoids')
plt.show()
```

```python
# prompt: show  me the code for standarzation for my whole data

from sklearn.preprocessing import StandardScaler


# Create a StandardScaler object
scaler = StandardScaler()


df_X = scaler.fit_transform(df)

df_X=pd.DataFrame(df_X)

# Now, df contains the standardized numerical features
print(df_X.head())
```

# Creating the cluster

```python
# Kmeans
from sklearn.cluster import KMeans
Kmeans=KMeans(n_clusters=4,n_init=20)

cluster1 = Kmeans.fit_predict(df_X)
cluster1
```

```python
Y2= pd.DataFrame(cluster1)
Y2.value_counts()
```

```python
Kmeans.inertia_
```

```python
#Checking for  elbow curve to be sure which K value is fit for the data

j=[]
for i in range(1,11):
  from sklearn.cluster import KMeans
  Kmeans=KMeans(n_clusters=i , n_init = 20)
  Kmeans.fit_predict(df_X)
  j.append(Kmeans.inertia_)

j
```

```python
#Showing the curve

plt.scatter(range(1,11),j)
plt.plot(range(1,11),j,color='red')
plt.show
```

# Aglomerative clustering

```python
##by single linkage method
Df1=df_X
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=4,
                                linkage= 'single')
cluster2=cluster.fit_predict(Df1)
cluster2
```

```python
Y= pd.DataFrame(cluster2)
Y.value_counts()
```

```python
#by complete linkage method
from sklearn.cluster  import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=4,
                                linkage= 'complete')
clusters3= cluster.fit_predict(Df1)
clusters3
```

```python
Y3= pd.DataFrame(clusters3)
Y3.value_counts()
```

```python
Df1['C_Cluster']=cluster.labels_
Df1.head(10)
```
